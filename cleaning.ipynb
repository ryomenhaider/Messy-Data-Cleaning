```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("retail_store_sales.csv")
```

```python
median_of_ppu = df["Price Per Unit"].median()
median_of_quantity = df["Quantity"].median()
median_of_total = df["Total Spent"].median()

print(f" Median of Price Per Unit: {median_of_ppu} \n Median of Quantity: {median_of_quantity} \n Median of Total Spent: {median_of_total}")

df = df.fillna(
    {
    "Price Per Unit": median_of_ppu,
    "Quantity": median_of_quantity,
    "Total Spent": median_of_total,
    "Discount Applied": "Unknown"
    }
)
```

```python
df.sample(10)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["Item"])
df.isnull().sum()
```

Every empty field is now cleaned


```python
df.duplicated(subset=["Customer ID", "Transaction ID"], keep=False)
df[df.duplicated(subset=["Customer ID", "Transaction ID"], keep=False)]
```

Changing dataTypes

```python
df.columns
```

```python
df = df.rename(
        columns={
            "Price Per Unit": "price_per_unit",
            "Total Spent": "total_spent",
            "Discount Applied": "discount_applied",
            "Transaction ID": "transaction_id",
            "Customer ID": "customer_id",
            "Item": "item",
            "Quantity": "quantity",
            "Transaction Date": "date",
            "Discount Applied": "discount_applied",
            "Location": "location",
            "Payment Method": "payment_method",
            "Category": "category"
        }
)
```

```python
cols = ["price_per_unit", "quantity", "total_spent"]
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce")
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```
